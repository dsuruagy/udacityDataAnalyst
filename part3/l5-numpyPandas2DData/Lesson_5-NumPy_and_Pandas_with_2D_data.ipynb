## Calculating Correlation

[Values of the Pearson Correlation](http://onlinestatbook.com/2/describing_bivariate_data/pearson.html)

[numpy.corrcoef](https://docs.scipy.org/doc/numpy/reference/generated/numpy.corrcoef.html)

```python
import pandas as pd

filename = 'nyc_subway_weather.csv'
subway_df = pd.read_csv(filename)

def correlation(x, y):
    '''
    Fill in this function to compute the correlation between the two
    input variables. Each input is either a NumPy array or a Pandas
    Series.
    
    correlation = average of (x in standard units) times (y in standard units)
    
    Remember to pass the argument "ddof=0" to the Pandas std() function!
    '''
    std_x = (x - x.mean())/x.std(ddof=0)
    std_y = (y - y.mean())/y.std(ddof=0)
    correlation = (std_x * std_y).mean()
    
    #corrcoef = np.corrcoef(x, y)
    #print corrcoef
    
    return correlation

entries = subway_df['ENTRIESn_hourly']
cum_entries = subway_df['ENTRIESn']
rain = subway_df['meanprecipi']
temp = subway_df['meantempi']

print correlation(entries, rain)
print correlation(entries, temp)
print correlation(rain, temp)

print correlation(entries, cum_entries)
```

## DataFrame applymap()

```python
import pandas as pd

# Change False to True for this block of code to see what it does

# DataFrame applymap()
if False:
    df = pd.DataFrame({
        'a': [1, 2, 3],
        'b': [10, 20, 30],
        'c': [5, 10, 15]
    })
    
    def add_one(x):
        return x + 1
        
    print df.applymap(add_one)
    
grades_df = pd.DataFrame(
    data={'exam1': [43, 81, 78, 75, 89, 70, 91, 65, 98, 87],
          'exam2': [24, 63, 56, 56, 67, 51, 79, 46, 72, 60]},
    index=['Andre', 'Barry', 'Chris', 'Dan', 'Emilio', 
           'Fred', 'Greta', 'Humbert', 'Ivan', 'James']
)

def letter_grade(grade):
    if grade >= 90 and grade <= 100:
        return 'A'
    elif grade >= 80 and grade < 90:
        return 'B'
    elif grade >= 70 and grade < 80:
        return 'C'
    elif grade >= 60 and grade < 70:
        return 'D'
    else:
        return 'F'

def convert_grades(grades):
    '''
    Fill in this function to convert the given DataFrame of numerical
    grades to letter grades. Return a new DataFrame with the converted
    grade.
    
    The conversion rule is:
        90-100 -> A
        80-89  -> B
        70-79  -> C
        60-69  -> D
        0-59   -> F
    '''
    return grades.applymap(letter_grade)

convert_grades(grades_df)
```

```python
def standardize_rows(df):
    '''
    Optional: Fill in this function to standardize each row of the given
    DataFrame. Again, try not to use apply().
    
    This one is more challenging than standardizing each column!
    '''
    mean_row = df.mean(axis='columns')
    std_row = df.std(axis='columns', ddof=0)
    
    return (df.sub(mean_row, axis='index')).div(std_row, axis='index')

standardize_rows(grades_df)
```

```python
data = subway_df.groupby(['day_week', 'rain'], as_index=False)['day_week', 'rain', 'ENTRIESn_hourly'].mean()

raining_data = pd.DataFrame(data[data['rain'] == 1])
raining_data.rename(columns={"ENTRIESn_hourly":"Raining"}, inplace=True)
del raining_data['rain']

print "\n"
non_raining_data = pd.DataFrame(data[data['rain'] == 0])
non_raining_data.rename(columns={"ENTRIESn_hourly":"Not raining"}, inplace=True)
del non_raining_data['rain']

join_data = raining_data.merge(non_raining_data, on='day_week', how='inner')
join_data.set_index('day_week', inplace=True) # Changing the index

%pylab inline
join_data.plot(title='Mean entries per day of week')
```

```python
#A scatterplot of subway stations with latitude and longitude as the x and y axes and ridership as 
#the bubble size
data_scatter = subway_df.groupby(['latitude','longitude'],as_index=False)['ENTRIESn_hourly'].sum()
data_scatter.plot.scatter(x='latitude', y='longitude', s=data_scatter['ENTRIESn_hourly']/9000,
                          alpha=0.5, title='Ridership by location')
```
